```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### BlackFriday Sales dashboard
- Plot 1: Total purchases by gender (Pie Chart)
- Plot 2: Total purchases by age group (Bar Chart)
- Plot 3: Purchases distribution by city category (Pie Chart)
- Plot 4: Occupation-based purchase insights (Bar Chart)

```python

# Load the data
data = pd.read_csv("BlackFriday.csv")

# Set the style for the dashboard
sns.set(style="whitegrid")

# Create subplots for the dashboard
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Total purchases by gender (Pie Chart)
gender_sales = data.groupby('Gender')['Purchase'].sum().reset_index()
axes[0, 0].pie(gender_sales['Purchase'], labels=gender_sales['Gender'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
axes[0, 0].set_title('Purchases Distribution by Gender')

# Plot 2: Total purchases by age group (Bar Chart)
age_sales = data.groupby('Age')['Purchase'].sum().reset_index()
sns.barplot(x='Age', y='Purchase', data=age_sales, ax=axes[0, 1], palette="coolwarm")
axes[0, 1].set_title('Total Purchases by Age Group')
axes[0, 1].set_ylabel('Total Purchase')

# Plot 3: Purchases distribution by city category (Pie Chart)
city_sales = data.groupby('City_Category')['Purchase'].sum().reset_index()
axes[1, 0].pie(city_sales['Purchase'], labels=city_sales['City_Category'], autopct='%1.1f%%', colors=['#ffcc99','#99ff99','#66b3ff'], startangle=90)
axes[1, 0].set_title('Purchases Distribution by City Category')

# Plot 4: Occupation-based purchase insights (Bar Chart)
occupation_sales = data.groupby('Occupation')['Purchase'].sum().reset_index()
sns.barplot(x='Occupation', y='Purchase', data=occupation_sales, ax=axes[1, 1], palette="rocket")
axes[1, 1].set_title('Total Purchases by Occupation')
axes[1, 1].set_ylabel('Total Purchase')

# Adjust layout
plt.tight_layout()

# Show the dashboard
plt.show()
```
